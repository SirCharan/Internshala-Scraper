```python
internshala.get('https://internshala.com/internships/matching-preferences')
for i in range(5,20):
#     internshala.get('https://internshala.com/internships/matching-preferences')
    stipend=internshala.find_elements_by_xpath('//span[@class="{}"]'.format("stipend"))[i].text.split(' ')[0]
    try:
        a=int(stipend)>5000
    except ValueError:
        if (str(stipend)=='Unpaid'):
            continue
        stipend=str(stipend.split('-')[0])
        a=int(stipend)>5000
    if (a):
        role=internshala.find_elements_by_xpath('//div[@class="{}"]'.format("heading_4_5 profile"))[i]
        if (role.text=='Digital Marketing'):
            print(i)
            company=internshala.find_elements_by_xpath('//div[@class="{}"]'.format("heading_6 company_name"))[i]
            Company.append(company.text)
            role.click()
            Role.append('Digital Marketing')
            Stipend.append(stipend)
            button=internshala.find_elements_by_xpath('//div[@class="{}"]'.format("buttons_container"))
            button[1].click()
            button2=internshala.find_elements_by_xpath('//div[@id="{}"]'.format("application_button"))
            button2[0].click()
            time.sleep(2)
            application_text=internshala.find_elements_by_xpath('//a[@class="{}"]'.format("copyCoverLetterTitle"))
            print(application_text)
            application_text[0].click()
            time.sleep(2)
            try:
                submit=internshala.find_elements_by_xpath('//div[@class="{}"]'.format("submit_button_container"))
                submit[0].click()
            except UnexpectedAlertPresentException:
                time.sleep(2)
                internshala.get('https://internshala.com/internships/matching-preferences')
                continue
        else:
            continue
    else:
        continue
```
